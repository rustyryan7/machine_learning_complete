```python
# Importing the neccessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
data = pd.read_csv('Position_Salaries.csv')
data.head()
```

```python
X = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
print(X)
print(y)
```

```python
# training linear regression model on dataset
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
regressor.predict([[7]])
```

```python
# training polynomial regression on dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X) # turning single feature into matrix of features
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
# Results using Linear regression with single feature
plt.scatter(X,y,color='red')
plt.plot(X,regressor.predict(X), color='blue')
plt.title('T or B Linear Regression model')
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualizing Polynomial Regression
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(X_poly), color='blue')
plt.title('T or B Polynomial Regression model')
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
# Predicting new salary using Linear regression
regressor.predict([[6.5]])
```

```python
# Predicting new salary using Polynomial regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
